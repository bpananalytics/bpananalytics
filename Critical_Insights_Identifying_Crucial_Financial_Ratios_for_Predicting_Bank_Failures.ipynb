### Comparative Analytics

```python
import numpy as np
from scipy.stats import shapiro, ks_2samp, levene, probplot
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Sample data
healthy_r1 = np.random.normal(healthy_firms['R1'])
failed_r1 = np.random.normal(failed_firms['R1'])

# Check for equal variance
levene_test = levene(healthy_r1, failed_r1)
print(f"Levene's test for equal variances: {levene_test}")

# Function to plot Q-Q plot with confidence intervals
def qq_plot_with_confidence(data, title):
    plt.figure()
    (osm, osr), (slope, intercept, r) = probplot(data, plot=None)

    # Q-Q plot
    plt.plot(osm, osr, 'o', markersize=5, label='Sample Data')

    # Line of best fit
    plt.plot(osm, slope * osm + intercept, 'r', label='Best Fit Line')

    # Confidence intervals
    z = 1.96  # for 95% confidence interval
    se = slope * np.sqrt(1/len(osm) + (osm - np.mean(osm))**2 / np.sum((osm - np.mean(osm))**2))
    lower_ci = slope * osm + intercept - z * se
    upper_ci = slope * osm + intercept + z * se

    plt.fill_between(osm, lower_ci, upper_ci, color='lightgray', label='95% Confidence Band')

    plt.xlabel('Theoretical Quantiles')
    plt.ylabel('Sample Quantiles')
    plt.title(title)
    plt.legend()
    plt.show()

# Normality check with Q-Q plot including confidence bands
qq_plot_with_confidence(healthy_r1, 'Q-Q plot R1 for healthy firms')
qq_plot_with_confidence(failed_r1, 'Q-Q plot R1 for failed firms')
```

```python
from scipy.stats import ttest_ind

# Perform t-test
ttest_result = ttest_ind(healthy_r1, failed_r1, equal_var=True)
print(f"T-test result: {ttest_result}")
```

**T-Test result suggests that Q1 is not significant**
