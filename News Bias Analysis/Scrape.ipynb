```python
# Save Townhall  Articles to CSV

import re
from bs4 import BeautifulSoup
import requests
import csv

headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/601.3.9 (KHTML, like Gecko) Version/9.0.2 Safari/601.3.9'
}
url = 'https://townhall.com/columnists/katie-pointer-baney/2024/10/16/harris-rhetorical-shift-on-guns-reflects-widespread-support-for-self-defense-n2646240'
response = requests.get(url, headers=headers)

# Parse the HTML content of the response
soup = BeautifulSoup(response.content, 'html.parser')

# TITLE

title_tag = soup.find('h1', {'id': 'post-title entry-title'})

# If the title tag is found, extract its text
if title_tag:
    print(title_tag.get_text())
else:
    # Fallback to the <title> tag in case <h1> is not found
    title_tag = soup.find('title')
    if title_tag:
        print(title_tag.get_text())
    else:
        print("Title not found")

#Author
author_tag = soup.find('div', class_='author').find('a').find('img')

author_name = author_tag['alt']

# Date
time_tag = soup.find('time', class_='date')

date_tag = soup.find('span', class_='date date-formatted')
original_date = date_tag['data-original-date']

# Image
container_div = soup.find('div', class_='post-header__img')
img_tag = container_div.find('img')['src']   


# figure caption
figcaption_tag = soup.find('figcaption')

# Extract the figcaption text if found
if figcaption_tag:
    figcaption_text = figcaption_tag.get_text(strip=True)  # Get the figcaption text and strip whitespace
    print(f"Figcaption: {figcaption_text}")
else:
    print("Figcaption not found")

    
# Body 
# Find the <div> tag with the class 'tds-content'
content_div = soup.find('section', class_='post-body')

text = ""

# If the content div is found, find all <p> tags within it
if content_div:
    paragraphs = content_div.find_all('p')  # Find all <p> tags
    for tag in paragraphs:
        text += tag.get_text()

    # Extract and print the text from each <p> tag
    #for i, p in enumerate(paragraphs, start=1):
    #    print(f"Paragraph {i}: {p.get_text(strip=True)}")  # Print each paragraph with a number
else:
    print("Content div not found")

# Cleaning Texts     
cleaned_text = re.sub(r'\bpic\.twitter\S*', '', text)
cleaned_text = re.sub(r'🔥', '', cleaned_text)
cleaned_text = re.sub(r'🚨', '', cleaned_text)
cleaned_text = re.sub(r'🤯', '', cleaned_text)
cleaned_text = re.sub(r'border@BretBaier', '', cleaned_text)
print(cleaned_text)


# Create a CSV file and write the data
with open('TownHall_guns.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    
    # Write the header
    writer.writerow(['Title', 'Author', 'Date', 'Image URL', 'Figcaption', 'Article Text'])
    
    # Write the article data
    writer.writerow([title_tag.get_text() , author_name, original_date, img_tag, figcaption_text, cleaned_text])

print("Done")
```
